```python
import os
import json

def rename_key(menu):
    i = 0
    for line in menu:
        if "price" in line and "itemsubtotal" not in line:
            line["itemsubtotal"] = line.pop("price")
            print(line)
            i += 1
    return i

        


file_path = "C:\\Users\\user\\OneDrive\\Documents\\MONASH SEM6\\FIT3164 - FYP\\JPG_BATCH_1\\edited2.json"
jsons = []

with open(file_path, 'r', encoding='utf-8') as f:
    for line in f:
        try:
            data = json.loads(line)
        except json.JSONDecodeError:
            print(f"Skipping invalid JSON")
            print(line)

with open(file_path, 'r', encoding='utf-8') as f:
    edit_counts = 0
    for line in f:
        data = json.loads(line)
        menu = None
        menu = data["ground_truth"]["gt_parse"]["menu"]
        if menu:
            edit_counts += rename_key(menu)
            jsons.append(data)
        else:
            print("THIS LINE DOESNT HAVE MENU")
            print(data)
print(str(edit_counts)+ " lines edited")
print(jsons)
```

```python
sorted_dataset = sorted(jsons, key=lambda x: x['id'])
with open("C:\\Users\\user\\Downloads\\1-80(new).json", "w") as f:
    for item in sorted_dataset:
        f.write(json.dumps(item) + "\n")
```

```python
import os
import json

file_path = "C:\\Users\\user\\OneDrive\\Documents\\MONASH SEM6\\FIT3164 - FYP\\New folder"
json_files = [f for f in os.listdir(file_path) if f.lower().endswith('.json')]
for j in json_files:
    jsons = []
    print(j)
    json_path = os.path.join(file_path, j)
    with open(json_path, 'r', encoding='utf-8') as f:
        for line in f:
            data = json.loads(line)
            data["ground_truth"] = json.dumps(data["ground_truth"])
            jsons.append(data)

    output_filename = j[:-5] + "_edited.json"
    output_file = os.path.join(file_path, output_filename)
    with open(output_file, 'w', encoding='utf-8') as out_f:
        for obj in jsons:
            out_f.write(json.dumps(obj) + '\n')
    print(f"saved edited file to {output_file}")


        
```
